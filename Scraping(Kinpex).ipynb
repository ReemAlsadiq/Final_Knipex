```python
def get_soup(url):  
    try:
        req = requests.get(url,verify=False)
        soup = BeautifulSoup(req.text)
        #print(soup)
        return soup
    except:
        pass
```

```python
df_soup = df['Product URL'].apply(lambda x: get_soup(x))
```

```python
def get_img(soup):
    try:
        img = soup.find('img',{'class': 'prod_zoom'}).get('src')
        #print(img)
        return img
    except:  
        pass
```
